```python
# Considering only numerical data 
df.data = df.iloc[:,1:]
df.data.head()
# Converting into numpy array
Win = df.data.values
Win
```

```python
# Normalizing the numerical data 
from sklearn.preprocessing import scale

df_norm=scale(Win)
df_norm
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit_transform(df_norm)
```

```python
pca_values
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(df_norm)
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(20,10))
dendrogram=sch.dendrogram(sch.linkage(df_norm,'ward'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3, linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(df_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
wine3=df.copy()
wine3['clustersid']=hclusters.labels_
wine3
```

```python
# K-Means Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Scree Plot / Elbow Curve
plt.plot(range(1, 6), wcss)
plt.title('Scree Plot / Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within-cluster sum of squares
plt.show()
```

```python
clusters3=KMeans(3, random_state=45).fit(df_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
df2=df.copy()
df2['clusters3id']= clusters3.labels_
df2
```
